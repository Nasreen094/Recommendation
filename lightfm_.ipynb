```python
import numpy as np
import scipy
import scipy.sparse as sp
import pandas as pd
from scipy.sparse import coo_matrix
```

```python
df=pd.read_csv("purchase_1127_1.csv",usecols=['CardNo','productname'])
data = df.dropna()
data.columns=['user','item']
data = data.copy()
user_item=pd.crosstab(index=data['user'], columns = data['item'])
user_item = user_item.reindex()
```

```python
replace=user_item.replace(0,-1)
```

```python
sparse=scipy.sparse.csr_matrix(replace)
```

```python
from lightfm import LightFM
model=LightFM(loss="warp")
model.fit(sparse, epochs=20)
```

```python
def sample_recommendation(model,data,user_id):
    n_users, n_items = data.shape
    known_positives=[user_item.iloc[user_id].axes[0][i] for i in range(9) if user_item.iloc[user_id][i] == 1]
    print("known positives")
    for x in known_positives:
        print("     %s" % x)
        
    scors=model.predict(user_id,np.arange(n_items))
    print scors
    top_items=data.columns[np.argsort(-scors)]
    print("recommended")
    for x in top_items[:3]:
        print("     %s" % x)
```

```python
sample_recommendation(model,user_item,200)
```
